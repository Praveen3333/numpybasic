````python
### Q1. Is there any difference in the data type of variables `list_` and `array_list`? If there is then write a code to print the data types of both the variables.

Yes, there is a difference. `list_` is a Python list, while `array_list` is a NumPy array. We can print their data types as follows:

```python
import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

print("Data type of list_:", type(list_))
print("Data type of array_list:", type(array_list))
```

### Q2. Write a code to print the data type of each and every element of both the variables `list_` and `array_list`.

```python
print("Data types of elements in list_:")
for element in list_:
    print(type(element))

print("Data types of elements in array_list:")
for element in array_list:
    print(type(element))
```

### Q3. Considering the following changes in the variable, `array_list`:
`array_list = np.array(object = list_, dtype = int)`
Will there be any difference in the data type of the elements present in both the variables, `list_` and `array_list`? If so then print the data types of each and every element present in both the variables, `list_` and `array_list`.

Yes, there will be a difference. The elements in `list_` will remain as strings, but the elements in `array_list` will be converted to integers.

```python
array_list = np.array(object=list_, dtype=int)

print("Data types of elements in list_:")
for element in list_:
    print(type(element))

print("Data types of elements in array_list after conversion:")
for element in array_list:
    print(type(element))
```

### Q4. Write a code to find the following characteristics of variable, `num_array`:
(i) shape
(ii) size

```python
num_list = [[1, 2, 3], [4, 5, 6]]
num_array = np.array(object=num_list)

print("Shape of num_array:", num_array.shape)
print("Size of num_array:", num_array.size)
```

### Q5. Write a code to create a numpy array of 3x3 matrix containing zeros only, using a numpy array creation function.
[Hint: The size of the array will be 9 and the shape will be (3,3).]

```python
zero_matrix = np.zeros((3, 3))
print("3x3 matrix of zeros:\n", zero_matrix)
```

### Q6. Create an identity matrix of shape (5,5) using numpy functions.
[Hint: An identity matrix is a matrix containing 1 diagonally and other elements will be 0.]

```python
identity_matrix = np.identity(5)
print("5x5 identity matrix:\n", identity_matrix)
```
````
